1. Describe how the posterior predictive distribution is created for mixture models. \
To derive the posterior predictive distribution applied on mixture models, we need to first integrate over all the component parameters. In other words, we need to integrate over all possible values of the component's parameters given the data or obtain the component's posterior distribution. Then, we need to introduce weights into the posterior distribution from the components because  new datapoints aren't guaranteed to be generated by a specific component. Finally, we need to sum the weighted distributions across all the components to derive the posterior predictive distribution.

3. Describe how to do a regression of y on X but X has some missing values and how to perform a Bayesian analysis without throwing away the rows with missing values in X. \


To perform linear regression with missing values, we need to predict these missing values based on available values. Based on the assumptions about the patterns of these missing values, we can classify them into three categories including missing completely at random, missing at random, and missing not at random. Missing completely at random suggests that the data missing doesn't correlate to the observed and unoberved parts of the data. Missing at random assumes that the reasons for missingness can be explained by function of the observed data. Missing not at random implies that the missingness is caused by something outside of the observed data, which makes it difficult for us to make estimations. Note that these three assumptions can't be verified. To deal with missing datas through Bayesian analysis, we let the missing values be unknown parameters and compute the posterior distribution of the unknown parameters.
